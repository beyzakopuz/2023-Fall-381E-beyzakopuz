```python
from urllib.request import urlopen
import xmltodict
```

```python
raw_data = []  # List to accumulate raw data
for year in range(1990, 2024):
    for roll_number in range(1, 1187):  # I checked all years and 2007 had the highest number of rolls 1186.
        # Constructing the URL for the roll call XML
        url = f'https://clerk.house.gov/evs/{year:04d}/roll{roll_number:03d}.xml'

        try:
            
            with urlopen(url) as url_obj:
                raw = xmltodict.parse(url_obj.read())
                # Add the raw data to the list
                raw_data.append(raw)

        except Exception as e:
            # Handle exceptions, for example, if the URL is not valid or the data is not available
            print(f"Error fetching data for {url}: {e}")
```

```python
import json
```

```python
# Specifying the local file path
file_path = r'C:\Users\beyza\OneDrive\Masast\MAT 381E\2023-Fall-381E-beyzakopuz\data\roll_calls.json'

# Saving all the raw data to a single JSON file
with open(file_path, 'w') as file:
    json.dump(raw_data, file, indent=2)

print(f'All raw data saved to {file_path}')
```
