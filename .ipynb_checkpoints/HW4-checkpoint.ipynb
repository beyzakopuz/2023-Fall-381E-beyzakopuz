```python
def remove_parts(book, start, end):
    pattern = re.compile(f'{re.escape(start)}(.*?){re.escape(end)}', re.DOTALL)
    match = re.search(pattern, book)
    if match:
        modified_book = match.group(1)
    return modified_book
```

```python
start_jane = '_April_ 13_th_, 1848.'
end_jane = '*** END OF THE PROJECT GUTENBERG EBOOK JANE EYRE: AN AUTOBIOGRAPHY ***'
modified_jane = remove_parts(jane_eyre, start_jane, end_jane)
print(modified_jane)
```

```python
start_pride = '[Illustration: ·PRIDE AND PREJUDICE·'
end_pride = '*** END OF THE PROJECT GUTENBERG EBOOK PRIDE AND PREJUDICE ***'
modified_pride = remove_parts(pride_prejudice, start_pride, end_pride)
print(modified_pride)
```

# Q3 (City hopping)

#### 1. Pull the data on airports and routes. Clean it and put the correct names on columns. 

```python
from io import StringIO
import numpy as np
```

```python
url4 = "https://raw.githubusercontent.com/jpatokal/openflights/master/data/airports-extended.dat"
response4 = requests.get(url4)
airports_raw = response4.text
```

```python
url5 = "https://raw.githubusercontent.com/jpatokal/openflights/master/data/routes.dat"
response5 = requests.get(url5)
routes_raw = response5.text
```

```python
def get_df(string_data, columns):
    data = StringIO(string_data)
    df = pd.read_csv(data, names=columns)
    return df
```

```python
airport_columns = ['Airport ID', 'City', 'Country', 'IATA', 'ICAO', 'Latitude', 'Longitude', 'Altitude',
                  'Timezone', 'DST', 'Tz database timezone', 'Type', 'Source']
```

```python
airports = get_df(airports_raw, airport_columns).replace('\\N', np.nan).reset_index(drop=True)
```

```python
airports
```

```python
routes_columns = ['Airline', 'Airline ID', 'Source airport', 'Source airport ID', 'Destination airport',
                 'Destination airport ID', 'Codeshare', 'Stops', 'Equipment']
```

```python
routes = get_df(routes_raw, routes_columns).replace('\\N', np.nan)
```

```python
routes
```

#### 2. Merge the route dataset and the airport dataset so the combined version has route source and target airports has full airport names and cities instead of airport codes.

(I will be sharing the full session in my references.)

```python
# Merge based on source airport
merged_df = pd.merge(routes, airports, left_on='Source airport', right_on='IATA', how='left')
merged_df = merged_df.rename(columns={'City': 'Source City', 'Country': 'Source Country'})

# Merge based on destination airport
merged_df = pd.merge(merged_df, airports, left_on='Destination airport', right_on='IATA', how='left')
merged_df = merged_df.rename(columns={'City': 'Destination City', 'Country': 'Destination Country'})

# Drop unnecessary columns
merged_df = merged_df.drop(['IATA_x', 'IATA_y'], axis=1)
```

```python
merged_df
```

#### 3. Construct a graph where each node is a city, and two cities are connected by an edge if there is a flight between these cities.

```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G = nx.DiGraph()

edges = [(row['Destination City'], row['Source City']) for _, row in merged_df.iterrows()]
G.add_edges_from(edges)

pos = nx.spring_layout(G) 
nx.draw(G, pos, with_labels=True, font_weight='bold', node_size=700, node_color='Pink', 
        font_size=8, edge_color='gray', linewidths=0.5, arrowsize=10)

# Show the plot
plt.show()
```

#### 1. Construct a graph where the nodes are actors and actresses, and two actors/actresses are connected by a weighted edge if these people appeared in the same movie together.

I asked ChatGpt which datasets I should download for this task and I obtained the datasets. (Full session link in references.)

```python
import gzip
```

```python
url7 = r"https://datasets.imdbws.com/title.basics.tsv.gz"
url8 = "https://datasets.imdbws.com/title.principals.tsv.gz"
```

```python
def get_imdb(url):
    response0 = requests.get(url)
    if response7.status_code == 200:
        data_compressed = BytesIO(response0.content)
        with gzip.open(data_compressed, 'rt', encoding='utf-8') as file:
            columnnames = file.readline().strip().split('\t')
            df = pd.read_csv(file, sep='\t', names=columnnames)

    else:
        print(f"Failed to fetch data. Status code: {response0.status_code}")
    return df
```

```python
title_basics = get_imdb(url7)
```

```python
titlebasics = title_basics.replace('\\N', np.nan)
```

```python
titlebasics.head()
```

```python
title_principals = get_imdb(url8)
```

```python
titleprincipals = title_principals.replace('\\N', np.nan)
```

```python
titleprincipals.head()
```

```python
if titlebasics is not None and titleprincipals is not None:

    relevant_columns = ['tconst', 'nconst', 'category']
    titleprincipals = titleprincipals[relevant_columns]

    # Filter rows where category is actor or actress
    actors_actresses = titleprincipals[
        title_principals['category'].isin(['actor', 'actress'])
    ]

    # Merge with titlebasics to get movie details
    merged_imdb = pd.merge(
        actors_actresses,
        titlebasics[['tconst', 'primaryTitle']],
        on='tconst',
        how='inner'
    )
```
