```python
from collections import Counter

def most_absent_legislator_by_year(absent_dict):
    most_absent_legislators = {}

    for year in range(1990, 2024):
        legislators_for_year = absent_dict.get(year, [])  # Get the list of legislators for the current year

        if not legislators_for_year:
            continue  # Skip if there are no legislators for the current year

        legislator_names = [legislator['legislator']['#text'] for legislator in legislators_for_year]

        # Count the occurrences of each legislator for the current year
        legislator_counts = Counter(legislator_names)

        # Find the most absent legislator for the current year
        if legislator_counts:
            most_absent_legislator_name = legislator_counts.most_common(1)[0][0]

            # Find the state of the most absent legislator
            most_absent_legislator_state = next(
                (legislator['legislator']['@state'] for legislator in legislators_for_year
                 if legislator['legislator']['#text'] == most_absent_legislator_name), None
            )

            most_absent_legislators[year] = {
                'legislator': most_absent_legislator_name,
                'state': most_absent_legislator_state
            }

    return most_absent_legislators
```

This is the function I ended up using. This function was very hard for me to write. I only could give ChatGpt a few lines of code to make the structure of my data and how to reach certain things.

```python
def create_legislator_dataframe(data):
    rows = []

    for i in range(len(data)):
        # Check if the required keys exist and are not None
        if 'rollcall-vote' in data[i] and data[i]['rollcall-vote'] is not None \
            and 'vote-metadata' in data[i]['rollcall-vote'] and data[i]['rollcall-vote']['vote-metadata'] is not None \
            and 'action-date' in data[i]['rollcall-vote']['vote-metadata'] \
            and 'vote-data' in data[i]['rollcall-vote'] and data[i]['rollcall-vote']['vote-data'] is not None:
            
            year = data[i]['rollcall-vote']['vote-metadata']['action-date']

            for j in range(len(data[i]['rollcall-vote']['vote-data']['recorded-vote'])):
                if 'legislator' in data[i]['rollcall-vote']['vote-data']['recorded-vote'][j]:
                    legislator_info = data[i]['rollcall-vote']['vote-data']['recorded-vote'][j]['legislator']
                    
                    # Check if the required keys exist and are not None
                    if '@state' in legislator_info and '#text' in legislator_info and '@party' in legislator_info:
                        state = legislator_info['@state']
                        name = legislator_info['#text']
                        party = legislator_info['@party']
                        vote = data[i]['rollcall-vote']['vote-data']['recorded-vote'][j]['vote']

                        rows.append([year, state, name, party, vote])

    # Create a DataFrame
    df = pd.DataFrame(rows, columns=['Year', 'State', 'Legislator', 'Party', 'Vote'])

    # Add columns for Number of times voted and Number of times did not vote
    df['Times Voted'] = df['Vote'].apply(lambda x: 1 if x == 'Yea' or x == 'Nay' else 0)
    df['Times Not Voted'] = df['Vote'].apply(lambda x: 1 if x == 'Not Voting' else 0)

    # Group by Year, State, Legislator, and Party and sum the counts
    df_grouped = df.groupby(['Year', 'State', 'Legislator', 'Party']).agg({
        'Times Voted': 'sum',
        'Times Not Voted': 'sum'
    }).reset_index()

    return df_grouped
```

```python
legislator_df = create_legislator_dataframe(data)
```

```python
from datetime import datetime
def str_to_date(your_df):
    date_format = "%d-%b-%Y"
    for i in range(len(your_df)):
        data_string = your_df['Year'][i]
        data_date = datetime.strptime(data_string, date_format)
        your_df['Year'][i] = data_date
    return your_df
```

This function substracts start year from end year for each legislator and finds the serving times.

```python
def longest_serving(start_df, end_df):
    serving_years = []
    for i in range(len(start_df)):
        serving = end_df['End'][i] - start_df['Start'][i]
        names_list = end_df.index.to_list()
        serving_years.append([names_list[i], serving.days // 365])
    return serving_years
```

```python
def yea_and_nay(roll_calls):
    vote_list = []
    for i in range(len(roll_calls)):
        action_date = data[i]['rollcall-vote']['vote-metadata']['action-date']
        roll_call = roll_calls[i]['rollcall-vote']['vote-metadata']
        vote_type = roll_call.get('vote-type', None)
        if vote_type and vote_type.startswith('YEA-AND-NAY'):
            try:
                yea_nay = roll_calls[i]['rollcall-vote']['vote-metadata']['vote-totals']['totals-by-party']
                rep = yea_nay[0]
                dem = yea_nay[1]
                rep_yea = float(rep['yea-total'])
                rep_nay = float(rep['nay-total'])
                dem_yea = float(dem['yea-total'])
                dem_nay = float(dem['nay-total'])
                pol = (abs(rep_yea - dem_yea) + abs(rep_nay - dem_nay)) / (rep_yea + rep_nay + dem_yea + dem_nay)
                vote_list.append([action_date, pol])
            except Exception:
                pass
            
    return vote_list
```
