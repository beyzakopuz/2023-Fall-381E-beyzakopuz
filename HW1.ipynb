```python
# (I wrote this function with the help of ChatGPT. I included the full session in my references.)

def get_monthly_averages(temperature):
    keys = ['ALB', 'BIH', 'CYP', 'DZA', 'EGY', 'ESP', 'GRC', 'HRV', 'ISR', 'ITA', 'LBN', 'LBY', 'MAR', 'MCO', 'MLT', 'MNE', 'PSE', 'REU', 'SVN', 'SYR', 'TUN', 'TUR']
    monthly_averages = []
    for ticker in keys:
        country = temperature[ticker]
        temp_list = list(country.values())
        for month in range(12):
            monthly_sum = sum(temp_list[month::12])  # Sum every 12th element (same month over 70 years)
            count = len(temp_list[month::12])  # Number of years
            monthly_average = monthly_sum / count  # Calculate average
            monthly_averages.append(((ticker, month, monthly_average)))
    return monthly_averages
```

```python
# Calculate monthly averages for all countries
monthly_averages = get_monthly_averages(tas) # tas is a dictionary.
```

```python
# (I wrote this function with the help of ChatGPT. I included the full session in my references.)

def get_average_temperature_per_annum(temperature):
    keys = ['ALB', 'BIH', 'CYP', 'DZA', 'EGY', 'ESP', 'GRC', 'HRV', 'ISR', 'ITA', 'LBN', 'LBY', 'MAR', 'MCO', 'MLT', 'MNE', 'PSE', 'REU', 'SVN', 'SYR', 'TUN', 'TUR']
    average_temperature_per_annum = {}

    for ticker in keys:
        country = temperature[ticker]
        temp_list = list(country.values())
        annual_averages = []
        for year in range(70):
            annual_sum = sum(temp_list[year * 12 : (year + 1) * 12])  # Sum temperatures for each month of the year
            annual_average = annual_sum / 12  # Calculate average for the year
            annual_averages.append(annual_average)
        average_temperature_per_annum[ticker] = annual_averages

    return average_temperature_per_annum
```

```python
import yfinance as yf
import datetime
```

```python
startdate = datetime.datetime.strptime("2019-01-01", "%Y-%m-%d").date()
enddate = datetime.datetime.strptime("2022-12-31", "%Y-%m-%d").date()
```

```python
stock = yf.download(["GOOG", "META", "CSCO", "INTC", "AMD", "QCOM", "AAPL", "005930.KS", "1810.HK", "TSLA"], start = startdate, end = enddate, group_by='tickers')
```

Cleaning the data. I dropped the null values in each company's data with the function below.

```python
# (I had the main idea and the structure of the function, I asked ChatGPT to help me with the rest.
# I shared the full session below in my References list.)

def clean_company_data(dataframe):
    company = ["GOOG", "META", "CSCO", "INTC", "AMD", "QCOM", "AAPL", "005930.KS", "1810.HK", "TSLA"]
    dataframe_new = {}
    # Loop through each company ticker and drop null values for that specific column
    for i in range(len(company)):
        dataframe_new[company[i]] = dataframe[company[i]].dropna()
    
    # Return the cleaned DataFrame
    return dataframe_new
```

```python
cleaned_stock_data = clean_company_data(stock)
```

## 1.

This function takes the opening and closing values of each company and calculates the values.

```python
# (I had the main idea and the structure of the function. I wrote the function as best as I could,
# then I asked ChatGPT to help me with the rest.
# I shared the full session below in my References list.)

def get_change(stock):
    company = ["GOOG", "META", "CSCO", "INTC", "AMD", "QCOM", "AAPL", "005930.KS", "1810.HK", "TSLA"]
    results = {}  # Dictionary to store results for each company
    
    for ticker in company:
        results[ticker] = []
        for index, row in stock[ticker].iterrows():
            a = row['Open'] + row['Open'] * 0.025
            b = row['Open'] + row['Open'] * -0.025
            
            if (row['Close'] >= b) and (row['Close'] <= a):
                results[ticker].append(0)
            elif row['Close'] > a:
                results[ticker].append(1)
            else:
                results[ticker].append(-1)
    
    return results
```

```python
results = get_change(cleaned_stock_data)
```

```python
type(results)
```

This function gets the dates from the data as indices for the change results we found.

```python
#(ChatGPT wrote this function for me, I only modified a couple of things. 
# I shared the full session below in my References list.)

def create_results_dataframes(results, cleaned_stock_data):
    company_dataframes = {}  # Dictionary to store DataFrames for each company
    
    for ticker, values in results.items():
        # Extract dates from the original DataFrame (assuming 'Date' is the column name)
        dates = cleaned_stock_data[ticker].index
        
        # Create a DataFrame for the current company with dates as indices and change values as columns
        company_df = pd.DataFrame({'Date': dates, 'Values': values})
        
        # Set 'Date' column as index
        company_df.set_index('Date', inplace=True)
        
        # Add the DataFrame to the dictionary with the ticker symbol as the key
        company_dataframes[ticker] = company_df
    
    return company_dataframes
```

```python
CHANGE = create_results_dataframes(results, cleaned_stock_data) 
# This is a dictionary. I couldn't turn it to a dataframe because the number of the rows aren't the same for each company
# since I only worked with the non-null values.
```

## 2.

This function finds the longest continuous time intervals that 'change' values of the companies remain 1 and -1.

```python
#(ChatGPT wrote this function for me, I only modified a couple of things. 
# I shared the full session below in my References list.)

def find_longest_intervals(data_dict, target_values):
    longest_intervals = {}

    for ticker, df in data_dict.items():
        for target_value in target_values:
            longest_interval_start = None
            longest_interval_end = None
            current_interval_start = None
            max_interval_length = 0

            for index, row in df.iterrows():
                if row['Values'] == target_value:
                    if current_interval_start is None:
                        current_interval_start = index

                    current_interval_end = index
                else:
                    if current_interval_start is not None:
                        interval_length = (current_interval_end - current_interval_start).days + 1
                        if interval_length > max_interval_length:
                            max_interval_length = interval_length
                            longest_interval_start = current_interval_start
                            longest_interval_end = current_interval_end

                        current_interval_start = None

            if current_interval_start is not None:
                interval_length = (current_interval_end - current_interval_start).days + 1
                if interval_length > max_interval_length:
                    longest_interval_start = current_interval_start
                    longest_interval_end = current_interval_end

            if ticker not in longest_intervals:
                longest_intervals[ticker] = {}

            longest_intervals[ticker][target_value] = (longest_interval_start, longest_interval_end)

    return longest_intervals
```

```python
target_values = [1, -1]
longest_intervals = find_longest_intervals(CHANGE, target_values)
```

```python
# (I had the main idea and the structure of the function. I wrote the function as best as I could,
# then I asked ChatGPT to help me with the rest.
# I shared the full session below in my References list.)


def get_difference(dictionary):
    company = ["GOOG", "META", "CSCO", "INTC", "AMD", "QCOM", "AAPL", "005930.KS", "1810.HK", "TSLA"]
    differences = {} 
    
    for i in company:
        differences[i] = []
        # Start iterating from the second row (index 1)
        for index in range(1, len(dictionary[i])):
            # Get the opening price of the current row
            current_open = dictionary[i]['Open'].iloc[index]
            
            # Get the opening price of the previous day (shifted one row up)
            previous_open = dictionary[i]['Open'].iloc[index - 1]
            
            # Calculate the thresholds
            lower_threshold = previous_open * -0.025 + previous_open
            upper_threshold = previous_open * 0.025 + previous_open
            
            # Determine the difference based on the relationship with the previous day's opening price
            if (current_open <= upper_threshold) and (current_open >= lower_threshold):
                differences[i].append(0)
            elif current_open >= upper_threshold:
                differences[i].append(1)
            else:
                differences[i].append(-1)
    return differences
```

```python
differences = get_difference(cleaned_stock_data)
```

```python
# This is the same function as the one I used to get the CHANGE dataframes.
# I only made a few litte modifications to fit this question (with the help of chatGPT.)

def create_differences_dataframes(differences, cleaned_stock_data):
    company_dataframes = {}  # Dictionary to store DataFrames for each company
    
    for ticker, values in differences.items():
        # Exclude the first row from cleaned_stock_data to match differences length
        stock_data_subset = cleaned_stock_data[ticker][1:]
        
        dates = stock_data_subset.index
        
        # Create a DataFrame for the current company with dates as indices and change values as columns
        company_df = pd.DataFrame({'Date': dates, 'Values': values})
        
        # Set 'Date' column as index
        company_df.set_index('Date', inplace=True)
        
        # Add the DataFrame to the dictionary with the ticker symbol as the key
        company_dataframes[ticker] = company_df
    
    return company_dataframes
```

```python
DIFFERENCE = create_differences_dataframes(differences, cleaned_stock_data)
# This is a dictionary. I couldn't turn it to a dataframe because the number of the rows aren't the same for each company
# since I only worked with the non-null values.
```
